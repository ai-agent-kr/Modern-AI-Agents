```python
import os

# .streamlit 폴더 생성
os.makedirs('.streamlit', exist_ok=True)

# secrets.toml 파일 생성 및 API 키 저장
openai_api_key = "여러분이 오픈 AI 홈페이지에서 발급받은 키 값"

secrets_content = f"""[general]
openai_api_key = "{openai_api_key}"
"""

with open('.streamlit/secrets.toml', 'w') as f:
    f.write(secrets_content)

print("✅ .streamlit/secrets.toml 파일이 생성되었습니다!")
print(f"📁 위치: {os.path.abspath('.streamlit/secrets.toml')}")
```
